### **Q2 Perform a matrix multiplication on the tensor from Q1 with another random tensor with shape `(1, 7)`**

```python
tensor_A = torch.rand(7,7)
tensor_B = torch.rand(1,7)

# Matrix multiplication
# transpose B from (1,7) to (7,1)

result = torch.matmul(tensor_A, tensor_B.T)


print("tensor_A shape:", tensor_A.shape)
print("tensor_B shape:", tensor_B.shape)
print("tensor_B.T shape:", tensor_B.T.shape)
print("Result shape:", result.shape)
print("\nResult:\n", result)
```

### **Q3 Set the random seed to `0` and do Q1 & Q2 over again**

```python
import torch

# Set random seed to 0
torch.manual_seed(0)

# Q1 - Random tensor of shape (7, 7)
tensor_A = torch.rand(7, 7)

# Q2 - Random tensor of shape (1, 7) and matrix multiplication
tensor_B = torch.rand(1, 7)
result = torch.matmul(tensor_A, tensor_B.T)

print("tensor_A:\n", tensor_A)
print("\ntensor_B:\n", tensor_B)
print("\nResult of matmul (tensor_A @ tensor_B.T):\n", result)
print("\nResult shape:", result.shape)
```


### **Q5 Create two random tensors of shape `(2, 3)` and send them both to the GPU. Set `torch.manual_seed(1234)` when creating the tensors (this doesn't have to be the GPU random seed)**

```python
import torch

# Q5 - Setup
torch.manual_seed(1234)
device = "cuda" if torch.cuda.is_available() else "cpu"

tensor_A = torch.rand(2, 3).to(device)
tensor_B = torch.rand(2, 3).to(device)

# Q6 - Matrix multiplication (transpose B to match inner dims)
# (2, 3) @ (3, 2) -> (2, 2)
output = torch.matmul(tensor_A, tensor_B.T)

print(f"tensor_A shape: {tensor_A.shape}")
print(f"tensor_B shape: {tensor_B.shape}")
print(f"tensor_B.T shape: {tensor_B.T.shape}")
print(f"\nOutput:\n {output}")
print(f"Output shape: {output.shape}")
```


### **Q10 Make a random tensor with shape `(1, 1, 1, 10)` and then convert it to a tensor of shape `(10)`. Set the seed to `7` when you create it and print out the first tensor and it's shape as well as the second tensor and it's shape**
