
### **Q5 Create two random tensors of shape `(2, 3)` and send them both to the GPU. Set `torch.manual_seed(1234)` when creating the tensors (this doesn't have to be the GPU random seed)**

```python
import torch

# Set seed (for reproducibility)
torch.manual_seed(1234)

# Create tensors on CPU
tensor1 = torch.rand(2, 3)
tensor2 = torch.rand(2, 3)

# Check if GPU is available
device = "cuda" if torch.cuda.is_available() else "cpu"

# Move tensors to GPU
tensor1 = tensor1.to(device)
tensor2 = tensor2.to(device)

print("Device:", device)
print("Tensor 1:\n", tensor1)
print("Tensor 2:\n", tensor2)
```


### **Q6 Perform a matrix multiplication on the tensors you created in Q5**

```python
import torch

torch.manual_seed(1234)

device = "cuda" if torch.cuda.is_available() else "cpu"

tensor1 = torch.rand(2, 3, device=device)
tensor2 = torch.rand(2, 3, device=device)

# Transpose tensor2
tensor2_T = tensor2.T

# Matrix multiplication
result = torch.matmul(tensor1, tensor2_T)

print("Result shape:", result.shape)
print(result)
```
